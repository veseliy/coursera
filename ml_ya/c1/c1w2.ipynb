```python
word_index = dict()
i = 0
for sentence in sentences:
    for word in sentence:
        if word not in word_index:
            word_index[word] = i
            i += 1
```

```python
matrix = []
for sent_i in xrange(0, len(sentences)):
    matrix.append([0 for x in word_index])
    
    for word in sentences[sent_i]:
        word_i = word_index[word]
        matrix[sent_i][word_i] += 1
```

```python
import math
import numpy as np
from scipy.linalg import solve
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
#f(x) = sin(x / 5) * exp(x / 10) + 5 * exp(-x / 2)
def sub_function(x):
    y=math.sin(float(x)/5)*math.exp(float(x) / 10)+ 5 * math.exp(-float(x) / 2)
    return y
```

```python
sub_function(1)
```

```python
a = np.array([[1, 1], [1, 15]])
b = np.array([sub_function(1), sub_function(15)])
```

```python
def sub_function_reg(w_0,w_1,x):
    y=[w_0+w_1*i for i in x]
    return y
```

```python
x=np.arange(0.0, 50.0, 0.1)
y_pred=sub_function_reg(w_0,w_1,x)
y=[sub_function(i) for i in x]
```
