```python
#Импорт библиотек
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import math
from scipy.stats import gamma
%matplotlib inline
```

```python
a = 1.99 #параметр гамма распределения
```

```python
r = gamma.rvs(a, size=1000) #гененирую выборку из 1000 элементов
```

```python
x = np.linspace(gamma.ppf(0.01, a), gamma.ppf(0.99, a), 100) #линейное-пространство для теоретического распределения
```

```python
rv = gamma(a) #создаю теоретическое распределение
```

```python
plt.hist(r, normed=True)
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
plt.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
#рисуночки
```

```python
nump = [] #список для средних значений
```

```python
mean, var, skew, kurt = gamma.stats(a, moments='mvsk')#центральные моменты первого порядка
```

# Генерация тестовой выборки и визуализация

```python
n = [5] #volume
k = 1000 #numbers of choice
for i in n:
    for j in range(0,k):
        r = gamma.rvs(a, size=i)
        mean = r.mean()
        nump.append(mean)
    plt.hist(nump, normed = True,bins = 10)  
x_mean = mean #выборочное среднее по цпт
x_std = math.sqrt(var/ (n[0])) #стд по цпт
norm_rv = sts.norm(x_mean, x_std) #нормальное распределение
x = np.linspace(0,4,100)
pdf = norm_rv.pdf(x)
plt.title('Graphic 1')
plt.plot(x, pdf, label='theoretical pdf', alpha=0.5) #вывод нормального распределения
plt.ylabel('fraction of samples')
plt.xlabel('$x$')
nump = []
```

## same for n =10
