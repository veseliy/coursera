```python
import numpy as np
import scipy.stats as stat
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#функция распределения парето с параметром 5 https://en.wikipedia.org/wiki/Pareto_distribution
#https://docs.scipy.org/doc/scipy-0.19.1/reference/generated/scipy.stats.pareto.html
func = stat.pareto(b=5)
#Случайная выборка из 1000 значений
r = func.rvs(1000)
#Точки для построения плотности распределения
x = np.linspace(1,4,1000)
#Плотность распределения от выборки
pdf = func.pdf(x)
#гистрограмма выборки и теоретическая функция распределения
plt.plot(x, pdf, label='PDF')
plt.hist(r, bins = 40, range=(1,4), normed=True)
plt.ylabel('$f(x)$/number of samples')
plt.xlabel('$x$')
plt.legend(loc='best')
```

```python
n=5
b=5
vib5=np.array([])#Массив выборочных средних
i=1
for i in range (1, 1000):#цикл генерации случайных выборок количества n из функции
    vib = func.rvs(n)#генерация выборки
   # print(vib)
    vs=sum(vib)/n#определение выборочного среднего
    vib5=np.append(vib5,vs)#добавление его в массив
#Параметры нормального распределения описывающие массив 1000 в.с. при n=5
E=(1*b)/(b-1)#среднее
D=(E**2)*(b/(b-2))#дисперсия
D5=D/n #дисперсия для нормального распределения приближающего выборку средних
sigma5=math.sqrt(D5)#сигма для нормального распределения выборки средних
#print (E,D,D5,sigma5)
norm5 = stat.norm(E, sigma5)#нормальное распределение с расчетными характеристиками Парето
pdf5 = norm5.pdf(x[:500])#получаем плотность распределения
#строим гистограмму массива средних при n=5 и плотность описывающую этот массив нормального распределения
plt.plot(x[:500], pdf5, label='PDF(n=5)')
plt.hist(vib5, bins =80, range=(0,5), normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
plt.legend(loc='best')
```

```python
# аналогично для n = 50
n=50
b=5
vib50=np.array([])
i=1
for i in range (1, 1000):
    vib = func.rvs(n)
    vs=sum(vib)/n
    vib50=np.append(vib50,vs)

E=(1*b)/(b-1)
D=(E**2)*(b/(b-2))
D50=D/n
sigma50=math.sqrt(D50)
norm50 = stat.norm(E, sigma50)
#print (E,D,D50,sigma50)
pdf5 = norm50.pdf(x[:300])
plt.plot(x[:300], pdf5, label='PDF(n=50)')
plt.hist(vib50, bins =80, range=(1,2), normed=True)
plt.ylabel('number of samples')
plt.xlabel('$x$')
plt.legend(loc='best')
```
